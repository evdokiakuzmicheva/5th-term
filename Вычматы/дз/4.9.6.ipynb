Покажем, что уравнение $ x + 0.5 \sin x + a = 0 $ имеет единственный корень  $\forall a \in \mathbb{R}$.  
$ f(x) = x + 0.5 \sin x + a\\
f'(x) = 1 + 0.5 \cos x > 0 \Rightarrow$  
f монотонно строго возрастает $\Rightarrow$ единственный корень  
$ \lim\limits_{x\to -\infty} f(x) = -\infty,\; \lim\limits_{x\to \infty} f(x) = \infty$

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def f(x, A):
    return x + 0.5 * np.sin(x) + A

def image(A):
    fig, ax = plt.subplots()
    x, f_ = np.linspace(-10, 10, num = 20), []
    f_ = f(x, A)
    ax.plot(x, f_, '.')
    ax.grid()
    plt.show()
    
# метод дихотомии
def solve(a, b, A):
    c = (a + b) / 2
    eps = 10 ** (-3)
    while (abs(f(c, A)) > eps):
        c = (a + b) / 2
        if f(a, A) * f(c, A) < 0:
            a, b = a, c
        elif f(b, A) * f(c, A) < 0:
            a, b = b, c
        else:
            print(round(c, 3))
            return
    print(round(c, 3))
```

### $a = 1$

```python
image(1)
a, b = -2.5, 0
solve(a, b, 1)
```

### $a = -1$

```python
image(-1)
a, b = 0, 1
solve(a, b, -1)
```

### $a = 2$

```python
image(2)
a, b = -2.5, 0
solve(a, b, 2)
```

### $a = -2$

```python
image(-2)
a, b = 0, 2.5
solve(a, b, -2)
```

### $a = 3$

```python
image(3)
a, b = -5, 0
solve(a, b, 3)
```

```python
image(-3)
a, b = 0, 5
solve(a, b, -3)
```

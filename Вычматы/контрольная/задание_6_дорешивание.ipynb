```python
import numpy as np
from matplotlib import pyplot as plt
from sympy import symbols, init_printing, sin, cos, exp
init_printing()
```

```python
def f(x):
    return np.sin(x ** 2) / x ** 2

x = np.linspace(1 / 2000, 1000, 1000)

#  
def rectangle(x):
    I = 0
    for i in range(len(x) - 1):
        h = x[i + 1] - x[i]
        x_ = (x[i] + x[i + 1]) / 2
        I += f(x_) * h
    return I

print((rectangle(x) + 0.0005).round(3))
```

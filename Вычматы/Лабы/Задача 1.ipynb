```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def F(x):
    return x ** 2 + np.tan(x) ** 2 - 1

eps = 10 ** (-6)
fig, ax = plt.subplots()
x = np.linspace(-1, 1, 30)
ax.plot(x, F(x))
ax.grid()
plt.show()
```

```python
# Метод дихотомии
def solve(a, b):
    c = (a + b) / 2
    eps = 10 ** (-6)
    while (abs(F(c)) > eps):
        c = (a + b) / 2
        if F(a) * F(c) < 0:
            a, b = a, c
        elif F(b) * F(c) < 0:
            a, b = b, c
        else:
            print(round(c, 3))
            return
    return round(c, 6)
    
print('x1 =', solve(-1, 0))
print('x2 =', solve(0, 1))
```

```python
import numpy as np
from sympy import Eq, symbols, init_printing, simplify, Matrix, sin, exp
from matplotlib import pyplot as plt

init_printing()
```

#### Вычисление функции u(t) = sin t на отрезке [0, 1]

```python
u = 0
n = 5

dt = 10 ** (-3)
t = symbols('t')
for i in range(1, n + 1):
    if i % 2 == 1:
        u += (-1) ** (int((i + 1) / 2) + 1) * t ** i / np.math.factorial(i)
print('Разложение функции u(t) = sin t в ряд Тейлора:')
display(u)
eps = abs(u - sin(t))

l = np.linspace(0, 1, 100)
arr = []
for i in range(len(l)):
    arr.append(eps.subs({t: l[i]}))

fig, ax = plt.subplots()
ax.plot(l, arr)
plt.xlabel('t')
plt.ylabel('отклонение')
ax.grid()
plt.show()

print('Из графика видно, что отклонение рассчитанного значения от реального увеличивается в направлении от 0 к 1.\nМаксимальное отклонение в точке t = 1 составило', np.array(arr).max().round(4), '.')
print('Было подобрано минимальное n = 5 так, чтобы эта максимальная ошибка не превышала заданную в условии точность 0.001.')
```

#### Разложение функций не в окрестности 0
Если надо разложить функцию на отрезке [10, 11] --> в окрестности не 0, применяем формулу:
$$ u(t) = u(t_0) + u'(t_0) (t - t_0) + ... + \frac{u^{(n)}(t_0)}{n!} (t - t_0)^n, $$
где $ t_0 = 10 $.

#### Вычисление функции u(t) = sin t на отрезке [10, 11]

```python
u = 0
n = 6

dt = 10 ** (-3)

def der_sin(p):
    while p > 3:
        p -= 4
    if p == 0:
        return np.sin(10)
    if p == 1:
        return np.cos(10)
    if p == 2:
        return -np.sin(10)
    if p == 3:
        return -np.cos(10)
    
for i in range(n):
    u += der_sin(i) * (t - 10) ** i / np.math.factorial(i)
print('Разложение функции u(t) = sin(t) на [10, 11] в ряд Тейлора:')
display(u)
eps = abs(u - sin(t))

l = np.linspace(10, 11, 100)
arr = []
for i in range(len(l)):
    arr.append(eps.subs({t: l[i]}))
    
fig, ax = plt.subplots()
ax.plot(l, arr)
plt.xlabel('t')
plt.ylabel('отклонение')
ax.grid()
plt.show()

print('Из графика видно, что отклонение рассчитанного значения от реального увеличивается в направлении от 10 к 11.\nМаксимальное отклонение в точке t = 11 составило', np.array(arr).max().round(4), '.')
print('Было подобрано минимальное n = 6 так, чтобы эта максимальная ошибка не превышала заданную в условии точность 0.001.')
```
